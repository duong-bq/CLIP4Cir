```python
import json

text_list = []
with open("fashionIQ_dataset\captions\cap.shirt.train.json") as f:
    dress_triplets = json.load(f)
    for item in dress_triplets:
        text_list.extend(item["captions"])

with open("fashionIQ_dataset\captions\cap.dress.train.json") as f:
    dress_triplets = json.load(f)
    for item in dress_triplets:
        text_list.extend(item["captions"])

with open("fashionIQ_dataset\captions\cap.toptee.train.json") as f:
    dress_triplets = json.load(f)
    for item in dress_triplets:
        text_list.extend(item["captions"])
```

```python
from collections import Counter
import re

# Combine the list into a single string and normalize it
text = " ".join(text_list).lower()

# Use regular expressions to find words and count them
words = re.findall(r'\w+', text)
word_count = Counter(words)
sorted_word_count = dict(sorted(word_count.items(), key=lambda item: item[1], reverse=True))

# Print the word count
for word, count in sorted_word_count.items():
    print(f"{word}: {count}")
```
